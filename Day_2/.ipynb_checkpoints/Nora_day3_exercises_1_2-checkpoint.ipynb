Exercise part 2 - Improving matmult.py

```python
import timeit 
%load_ext line_profiler

N = 250

def NxN_matrix(n):
    X = []
    for i in range(n):
        X.append([random.randint(0,100) for r in range(n)])
    return X
    
X = NxN_matrix(N)
    
%timeit NxN_matrix(N)
%memit NxN_matrix(N)

```

```python
#An improvement on the previous part

x = np.random.uniform(low=0,high=100,size=(N,N))
%timeit x
%memit x
```

```python
# Nx(N+1) matrix

def Nx_matrix(n):
    Y = []
    for i in range(n):
        Y.append([random.randint(0,100) for r in range(n+1)])
    return Y

Y = Nx_matrix(N)
    
%timeit Nx_matrix(N)
%memit Nx_matrix(N)
```

```python
#Improvement on Nx(N+1) matrix
y = np.random.uniform(low=0,high=100,size=(N,N+1))
%timeit y
print(np.shape(y))
```

```python
# result is Nx(N+1)

def Nx1(n):
    result = []
    for i in range(n):
        result.append([0] * (n+1))
    return result

result = Nx1(N)

%memit Nx1(N)
%timeit Nx1(N)
```

```python
#Improvement on previous part

result2 = np.zeros((N,N+1))
%memit result2
%timeit result2
```

```python
# iterate through rows of X

def iteration(X,Y,result):
   # iterate through rows of X
    for i in range(len(X)):
    # iterate through columns of Y
        for j in range(len(Y[0])):
        # iterate through rows of Y
            for k in range(len(Y)):
                result[i][j] += X[i][k] * Y[k][j]
    return result


%timeit iteration(X,Y,result)
%memit iteration(X,Y,result)

print(np.shape(iteration(X,Y,result)))
```

```python
#We get an improvement of a factor of 10^8 !!!
result1 = np.matmul(x,y)
%timeit result1
%memit result1
print(np.shape(result1))
        
```
